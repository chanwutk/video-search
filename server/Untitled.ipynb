```python
from videostore.videostore import VideoStore, preprocess
```

```python
import cv2
from PIL import Image
import numpy as np
```

```python
cam = cv2.VideoCapture('./data/ElFuente2_h264.mp4')
```

```python
fps = cam.get(cv2.CAP_PROP_FPS)
```

```python
fps
```

```python
images = []
```

```python
while(True):
    ret, frame = cam.read()
    if not ret: break
    image = Image.fromarray(frame[:,:,::-1], 'RGB')
    images.append(preprocess(image))
```

```python
image_input = np.stack(images)
```

```python
store = VideoStore('./video-data')
```

```python
store.add('kid', fps, image_input)
```

```python
store.list_videos()
```

```python
store.get_videos(['kid'])
```

```python
store.probability('A kid smiling', ['kid'])
```
